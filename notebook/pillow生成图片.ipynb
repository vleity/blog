```python
from PIL import Image, ImageDraw, ImageFont
import numpy as np
```

## 生成一张图片

```python
# 图片宽度
width = 700
# 图片高度
height = 110
# 字体
font = ImageFont.truetype("/usr/share/fonts/chinese/msyh.ttc", size=48)

# 创建一张图片
im = Image.new('RGB', (width, height), (250, 250, 250))
# 创建绘图对象
draw = ImageDraw.Draw(im)
# 添加边框
draw.line([(0,0),(width,0),(width,height),(0,height),(0,0)],width=8,fill=(230, 230, 230))
# 添加文字
draw.text((48, 20),'上善若水，水利万物而不争!', font=font, fill=(80, 80, 80))

im
```

```python
# 图片宽度
width = 700
# 图片高度
height = 110
# 字体
font = ImageFont.truetype("/usr/share/fonts/chinese/msyh.ttc", size=48)

# 创建一张图片
im1 = Image.new('RGB', (width, height), (250, 250, 250))
# 创建绘图对象
draw = ImageDraw.Draw(im1)
# 添加边框
draw.line([(0,0),(width,0),(width,height),(0,height),(0,0)],width=8,fill=(230, 230, 230))
# 添加文字
draw.text((48, 20),'冰，水为之，而寒于水！', font=font, fill=(80, 80, 80))

im1
```

## 图片缩放

```python
def resize(im,factor):
    """
    图片等比缩放
    """
    return im.resize((int(im.size[0]*factor),int(im.size[1]*factor)),Image.ANTIALIAS)
```

```python
resize(im,0.5)
```

## 转二进制IO，等价于open('xxx.png','rb')

```python
import io
def image_to_io(im):
    """
    Image转bytesio
    """
    tmp = io.BytesIO()
    im.save(tmp,format('png'))
    return tmp.getvalue()
```

```python
image_to_io(im)[:10]
```
